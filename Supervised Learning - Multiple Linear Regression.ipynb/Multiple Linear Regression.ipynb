Multiple Linear Regression

```python
X = data.iloc[:,[0,1,2]].values # All the rows, 3 types of Sales. Multiple columns.
X[1:4]
```

```python
y = data.iloc[:,3].values # Sales
y = y.reshape(-1,1)
X = X.reshape(-1,3) # You can also use lowercase x

# create an empty Linear regression model Like below and give it a good variable name
advertising_model = LinearRegression()

# to create the model, we use fit(x,y)
advertising_model.fit(X,y)

print('Intercept: \n', advertising_model.intercept_)
print('Coefficients: \n', advertising_model.coef_)
```

Training and Test Samples

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25, shuffle=True)

print ("Training data:",X_train.shape, y_train.shape) #Seen data. Observations, Number of columns.
print ("Test data:",X_test.shape, y_test.shape) # Unseen data. Observations, Number of columns.
```

Using a 0.25 test size split our data of 200 items into a training set of 150 observations and three columns,
and a test set of 50 observations.

We can use this to evaluate our model.

Now we'll fit the model on the training data:

```python
# fit a model
lm = LinearRegression()
# Fit model on training data
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
```

```python
predictions[0:5] # 1 to # 4
```

```python
## Plot the errors
residuals = y_test-predictions # Y testing data minus our predictions
residuals = np.squeeze(residuals) # Squeeze it to 1D array

#ensure only +tive values are taken for yerr
res_absolute = np.abs(residuals)
res_absolute = np.squeeze(res_absolute)

# Why did this take me so Long to figure out?
```

```python
plt.errorbar(range(1,51), np.squeeze(y_test), yerr=res_absolute, fmt='.k', ecolor="red") 
# We know our test data is 50 values long hence range(1,51)
# Squeeze to 1 D array
# Y-error calculation . We calculate the error value of the absolute residuals (res_absolute)
# format, .k
# Error colour , red
```

The red lines indicate the error of the prediction from the actual observations. (How fat off the predictions are from the observed values.)

Sklearn's LinearRegression provides the R squared score for its models:

```python
print ("Score:", model.score(X_test, y_test))
```

This score is quite close to 1, which suggests that our model fits our observations quite well.
