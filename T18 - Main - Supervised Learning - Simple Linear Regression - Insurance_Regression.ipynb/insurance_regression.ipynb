```python
import pandas as pd # For reading and preprocessing data for use in scikit-learn.
import numpy as np # Allows you to work with multidimensional arrays and matrices. Itâ€™s perfect for scientific or mathematical calculations.
import seaborn as sns # Essential for Data visualization.
import matplotlib.pyplot as plt # Essential for Data visualization.
from sklearn.linear_model import LinearRegression # Involves solving a system of linear equations
%matplotlib inline
```

```python
file_path = r"\Users\tailb\Data Science\JUPYTER NOTEBOOK\insurance.csv"
insurance_data = pd.read_csv(file_path,delim_whitespace="") # delim_whitespace=0
insurance_data.head()
```

```python
#x = insurance_data.loc[:,"age"].values # All the rows, Column labeled age.
#y = insurance_data.loc[:,"charges"].values # All the rows, Column labeled charges.

# iloc() Purely integer-location based indexing for selection by position.
x = insurance_data.iloc[:,:1].values.round(0) # All the rows, Only 1st Column. Select Age Column
y = insurance_data.iloc[:,-1].values.round(0) # All rows, Last Column. Select charges Column

plt.scatter(x,y,color = 'c') # c is cyan colour
plt.xlabel("Age") # Independent Variable
plt.ylabel("Charges") # Dependent Variable
plt.show()
```

Fitting a Linear Regression Model.
.fit() prepares the data for prediction.

```python
# create an empty Linear regression model
insurance_model = LinearRegression()

# to create the model, we use fit(x,y)
insurance_model.fit(x,y) # iloc.() is recommended!
```

We want to make a prediction of the Y value based on whatever value X value we feed.
The prediction values will be our â€˜line of best fitâ€™.

```python
y_pred = insurance_model.predict(x)

plt.scatter(x,y,color = 'c') # c is cyan colour
plt.plot(x,y_pred,color = 'r') # Plots line of best fit.
plt.xlabel("Age") # Independent Variable
plt.ylabel("Charges") # Dependent Variable
plt.show()
```
